```python
# DataFrame = pd.DataFrame()
# #for i in range(1, 69):
# path = "C:/Users/Barca/Dropbox/xls copies/html/upd_oct22/acquisitions 002.htm"
# soup = BeautifulSoup(open(path, encoding = 'utf-8'),'html.parser')

# #собираем заголовки признаков
# header = []
# criteries = soup.find_all("div", {"class":"header-contents"})
# for c in criteries:
#     header.append(c.get_text())
# header = header[1:-1]
# header[0] = 'RowNumber'
# header.append('Org_Code')
# header.append('Acquirer_Code')
# header.append('Operation_Code')
    
# #собираем признаки
# cells = []
# rows = soup.find_all("grid-row")
# for row in rows:
#     y = row.find_all("a", {"class":"component--field-formatter field-type-identifier link-accent ng-star-inserted"}, href = True)
#     x = row.find_all("div", {"class":"result-number ng-star-inserted"}) + row.find_all("field-formatter") + y
#     cells.append(x)
# data = []
# for row in cells:
#     cells_in_row_array = []
#     for cells_in_row in row:
#         cells_in_row_array.append(cells_in_row.get_text())
#     cells_in_row_array[-3] = row[-3]['href']
#     cells_in_row_array[-2] = row[-2]['href']
#     cells_in_row_array[-1] = row[-1]['href']
#     data.append(cells_in_row_array)

# #заносим в датафрейм
# df = pd.DataFrame(data = data, columns = header)
# DataFrame = pd.concat([DataFrame, df], ignore_index=True, sort=False)
    
# DataFrame.to_csv('acquisitions_upd2.csv', encoding = 'utf-8-sig') #, index=False
# DataFrame


DataFrame = pd.DataFrame()
for i in range(1, 140):
    path = "C:/Users/Barca/Dropbox/xls copies/html/exits/acquisitions/" + str(f"{i:03d}") + ".html"
    soup = BeautifulSoup(open(path, encoding = 'utf-8'),'html.parser')

    #собираем заголовки признаков
    header = []
    criteries = soup.find_all("div", {"class":"header-contents"})
    for c in criteries:
        header.append(c.get_text())
    header = header[1:-1]
    header[0] = 'RowNumber'
    header.append('Org_Code')
    header.append('Acquirer_Code')
    header.append('Operation_Code')
    
    #собираем признаки
    cells = []
    rows = soup.find_all("grid-row")
    for row in rows:
        y = row.find_all("a", {"class":"component--field-formatter field-type-identifier link-accent ng-star-inserted"}, href = True)
        x = row.find_all("div", {"class":"result-number ng-star-inserted"}) + row.find_all("field-formatter") + y
        cells.append(x)
    data = []
    for row in cells:
        cells_in_row_array = []
        for cells_in_row in row:
            cells_in_row_array.append(cells_in_row.get_text())
        cells_in_row_array[-3] = row[-3]['href']
        cells_in_row_array[-2] = row[-2]['href']
        cells_in_row_array[-1] = row[-1]['href']
        data.append(cells_in_row_array)

    #заносим в датафрейм
    df = pd.DataFrame(data = data, columns = header)
    DataFrame = pd.concat([DataFrame, df], ignore_index=True, sort=False)
    
DataFrame.to_csv('acquisitions.csv', encoding = 'utf-8-sig') #, index=False
DataFrame
```

```python
for j in range(83, 84):
    DataFrame = pd.DataFrame()
    for i in range(1, 18):
        path = "C:/Users/Barca/Dropbox/xls copies/html/2007-2000/block" + str(f"{j:02d}") + "/" + str(f"{i:03d}") + ".html"
        soup = BeautifulSoup(open(path, encoding = 'utf-8'),'html.parser')

        #собираем заголовки признаков
        header = []
        criteries = soup.find_all("div", {"class":"header-contents"})
        for c in criteries:
            header.append(c.get_text())
        header = header[1:-1]
        header[0] = 'RowNumber'
        header.append('Org_Code')
        header.append('Round_Code')    

        #собираем признаки
        cells = []
        rows = soup.find_all("grid-row")
        for row in rows:
            y = row.find_all("a", {"class":"component--field-formatter field-type-identifier link-accent ng-star-inserted"}, href = True)
            x = row.find_all("div", {"class":"result-number ng-star-inserted"}) + row.find_all("field-formatter") + y
            cells.append(x)
        data = []
        for row in cells:
            cells_in_row_array = []
            for cells_in_row in row:
                cells_in_row_array.append(cells_in_row.get_text())
            cells_in_row_array[-2] = row[-2]['href']
            cells_in_row_array[-1] = row[-1]['href']
            data.append(cells_in_row_array)

        #заносим в датафрейм
        df = pd.DataFrame(data = data, columns = header)
        DataFrame = pd.concat([DataFrame, df], ignore_index=True, sort=False)

    DataFrame.to_csv("blocks 2007-2000/block" + str(f"{j:02d}") + ".csv", encoding = 'utf-8-sig') #, index=False
    DataFrame
```
