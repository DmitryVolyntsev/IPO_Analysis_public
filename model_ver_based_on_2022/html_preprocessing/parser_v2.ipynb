```python
# Importing the required modules 
import os
import sys
import pandas as pd
from bs4 import BeautifulSoup
```

```python
path1 = 'C:/Users/Barca/Downloads/Search_Results_Companies_&_Deals1.html'
path = 'C:/Users/Barca/Desktop/GitHub/IPO_Analysis/НИР_2020/Advanced_Search_Results_Companies_&_Deals1.html'
soup = BeautifulSoup(open(path),'html.parser')
#soup
```

```python
#собираем заголовки признаков
header = []
criteries = soup.find_all("div", {"class":"smart-caption smart-caption__static"})
for c in criteries:
    #try:
    header.append(c.get_text()) #c.find_all("span", {"class":"ellipsis smart-caption__text"})
    #except:
    #    continue
```

```python
#собираем названия компаний
cmpns = []
companies = soup.find_all("span", {"class":"entity-hover ellipsis entity-format__entity-profile"})
for company in companies:
    cmpns.append(company.get_text())

#собираем признаки
cells = []
rows = soup.find_all("div", {"class":"data-table__row"})
for row in rows:
    cells.append(row.find_all("div", {"class":"data-table__cell"}))
    #row.find_all("span", {"class":"ellipsis"})
data = []
for row in cells:
    cells_in_row_array = []
    for cells_in_row in row:
        cells_in_row_array.append(cells_in_row.get_text())
    data.append(cells_in_row_array)
data = data[28:]

#объединяем
for i in range(len(cmpns)):
    data[i].insert(0, cmpns[i])

#заносим в датафрейм
dataFrame = pd.DataFrame(data = data, columns = header)
dataFrame#.to_csv('info.csv')
```
