```python
#closed
DataFrame = restricted_list
DataFrame[' Closed Date '] = pd.to_datetime(DataFrame[' Closed Date '], errors='coerce')
DataFrame[' Last Funding Date '] = pd.to_datetime(DataFrame[' Last Funding Date '], errors='coerce')
DataFrame[' Founded Date '] = pd.to_datetime(DataFrame[' Founded Date '], errors='coerce')
DataFrame[' Exit Date '] = pd.to_datetime(DataFrame[' Exit Date '], errors='coerce')

DataFrame[' Total Funding Amount '] = DataFrame[' Total Funding Amount '].replace('[\,]', '', regex=True)
DataFrame[' Last Funding Amount '] = DataFrame[' Last Funding Amount '].replace('[\,]', '', regex=True)

DataFrame[' Currency '] = DataFrame[' Total Funding Amount '].str.extract('(\D+)', expand=True)
DataFrame[' Total Funding Amount '] = DataFrame[' Total Funding Amount '].str.extract('(\d+)', expand=True)
DataFrame[' Total Funding Amount '] = pd.to_numeric(DataFrame[' Total Funding Amount '], errors='coerce')
DataFrame[' Last Funding Amount '] = DataFrame[' Last Funding Amount '].str.extract('(\d+)', expand=True)
DataFrame[' Last Funding Amount '] = pd.to_numeric(DataFrame[' Last Funding Amount '], errors='coerce')
DataFrame.to_csv('closed.csv', encoding = 'utf-8-sig') #, index=False
DataFrame
```
